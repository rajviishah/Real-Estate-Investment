## **Kmeans**

K-means clustering is a simple learning algorithm that is used to solve clustering problems. It follows a simple procedure of classifying a given data set into a number of clusters, defined by the letter "k," which is fixed beforehand. The clusters are then positioned as points and all observations or data points are associated with the nearest cluster, computed, adjusted and then the process starts over using the new adjustments until a desired result is reached.

Reducing the dimensions of the data set using PCA.

```python
scaler =  StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)

pca = PCA(n_components = 2)
transformed_data = pca.fit_transform(scaled_data)
```

we can see that for 2 components only 25% of the variance is covered and we lose a lot of features.

```python
pca = PCA(n_components = 10)
transformed_data = pca.fit_transform(scaled_data)
np.cumsum(pca.explained_variance_ratio_ * 100)
```

Elbow Plot

The technique to determine K, the number of clusters.

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1, 15):
    kmeans=KMeans(n_clusters = i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

lets visualise for k = 3

```python
kmeans = KMeans(n_clusters = 3)
y_predicted = kmeans.fit_predict(df)
y_predicted
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=y_predicted)
plt.xlabel('PCA Feaure 1')
plt.ylabel('PCA Feature 2')
```

# **Gaussian Mixture**

A Gaussian Mixture is a function that is comprised of several Gaussians, each identified by k âˆˆ {1,â€¦, K}, where K is the number of clusters of our dataset. Each Gaussian k in the mixture is comprised of the following parameters:
1. A mean that defines its centre.
2. A covariance that defines its width. This would be equivalent to the dimensions of an ellipsoid in a multivariate scenario.
3. A mixing probability that defines how big or small the Gaussian function will be.

Importing library.

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components = 3).fit(df)
labels = gmm.predict(df)
plt.scatter(df["TotalBsmtSF"], df["SalePrice"], c=labels, s=40, cmap='viridis')
```
